```python
#1. 输入库
import numpy as np
import pandas as pd

#2. 打开d带中心数据
with open('My_Data.csv', encoding = 'UTF-8') as f:
    df = pd.read_csv(f, index_col = 0)  #index_col=0 告诉第一列是索引

df
```

```python
type(df)
```

```python
#设置自变量和因变量
X = df.drop('Average', axis = 1)
Del = df['Average']
```

```python
#对列遍历
#for i in X.columns:  #对列遍历
 #   X.loc[,i] -= Del
    
for i in X.columns:  #对列遍历
    for j in X.index:  #对行遍历
        X.loc[j,i] -= Del.loc[j]
```
